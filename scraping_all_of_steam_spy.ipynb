```python
list_of_columns = ["appid","Name","Developer","Publisher","Owners","Metascore","Genre(s)"]
#steam_spy_df = pd.DataFrame(columns=list_of_columns)

#u = 0

list_of_names = []
list_of_developers = []
list_of_publishers = []
list_of_metascores = []
list_of_owners = []
list_of_genres = []

for u in list_of_steam_ids:
    print("Trying #" + str(u))
    url = "https://steamspy.com/app/" + str(u)
    #url2 = "https://steamspy.com/app/10"
    print(url)
    r = requests.get(url, headers=header)
    soup = BeautifulSoup(r.text, 'html.parser')
    title_raw = soup.find("h3")
    title = title_raw.get_text()
    if title == "":
        #u = u + 1
        print("oops")
        list_of_names.append("n/a")
    else:
        list_of_names.append(title)
        developer = soup.find(text="Developer:")
        developer_next = developer.parent.findNext()
        developer_text = developer_next.get_text()
        list_of_developers.append(developer_text)
        publisher = soup.find(text="Publisher:")
        publisher_next = publisher.parent.findNext()
        publisher_text = publisher_next.get_text()
        list_of_publishers.append(publisher_text)
        try:
            metascore = soup.find('strong', text="Metascore:")
            metascore_next = metascore.next_sibling.strip()
            list_of_metascores.append(metascore_next)
        except AttributeError:
            print("No Metascore!")
            list_of_metascores.append("n/a")
        owners = soup.find('strong', text="Owners")
        owners_next = owners.next_sibling.strip()
        owners_next = owners_next[2:]
        list_of_owners.append(owners_next)
        genre = soup.find('strong', text="Genre:")
        genre_parent = genre.parent()
        i = 0
        end_of_genre_parent = len(genre_parent)
        end_of_line = False
        found_genre = False
        #list_of_genres = []
        genre2=[]
        for i in range(len(genre_parent)):
            #print(genre_parent[i])
            if "/genre" in str(genre_parent[i]):
                genre2.append(genre_parent[i].get_text())
                #list_of_genres_df.append(list_of_genres)
        list_of_genres.append(genre2)
        
#steam_spy_df = pd.DataFrame(columns=list_of_columns)
#steam_spy_df

steam_spy_df = pd.DataFrame()
steam_spy_df["App Id"] = list_of_steam_ids
steam_spy_df["Name"] = list_of_names
steam_spy_df["Developers"] = list_of_developers
steam_spy_df["Publishers"] = list_of_publishers
steam_spy_df["Metascores"] = list_of_metascores
steam_spy_df["Owners"] = list_of_owners
steam_spy_df["Genres"] = list_of_genres
```
